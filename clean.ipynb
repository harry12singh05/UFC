```python
df["win_by"]=df["win_by"].fillna(df["win_by"].mode()[0])
```

```python
df["Format"]=df["Format"].fillna(df["Format"].mode()[0])
```

```python
df.drop(columns=["Referee"],inplace=True)
```

```python
df.drop(columns=["location"],inplace=True)
```

```python
df.dropna(subset="Winner",inplace=True)
```

```python
df.drop(columns=["R_DOB"],inplace=True)
```

```python
df.drop(columns=["B_DOB"],inplace=True)
```

```python
df=df.fillna({
    "Fight_type":df["Fight_type"].mode()[0],
    "R_Stance":df["R_Stance"].mode()[0],
    "R_Height":df["R_Height"].median(),
    "R_Reach":df["R_Reach"].median(),
    "B_Stance":df["B_Stance"].mode()[0],
    "B_Height":df["B_Height"].median(),
    "B_Reach":df["B_Reach"].median(),
    "R_Age":df["R_Age"].median(),
    "B_Age":df["B_Age"].median()
},inplace=True)
```

```python
df["R_Weight"]=df["R_Weight"].str.extract(r"(\d+)").astype("float")
```

```python
df["R_Weight"]=df["R_Weight"].fillna(df["R_Weight"].median())
```

```python
df["B_Weight"]=df["B_Weight"].str.extract(r"(\d+)").astype("float")
```

```python
df["B_Weight"]=df["B_Weight"].fillna(df["B_Weight"].median())
```

```python
df.rename(columns={
    "R_Weight":"Red Weight",
    "B_Weight":"Blue Weight",
    "R_Height":"Red Height",
    "B_Height":"Blue Height",
    "R_Reach":"Red Reach",
    "B_Reach":"Blue Reach",
    "R_Age":"Red Age",
    "B_Age":"Blue Age",
    "R_Stance":"Red Stance",
    "B_Stance":"Blue Stance",
    "Fight_type":"Fight Type",
    "win_by":"Win By"
},inplace=True)
```
