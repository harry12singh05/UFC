```python
plt.style.use('ggplot')
```

```python
sns.set_theme(style="whitegrid")
sns.set_palette("Set2")
```

3.DASHBOARD

```python
weight_df=df[(df["Red Weight"]<=300) & (df["Blue Weight"]<=300)]

sns.histplot(weight_df["Red Weight"],
                    bins=10,
                 color="red",
                 alpha=0.5,
                 label="Red"
                 
                 )
sns.histplot(weight_df["Blue Weight"],
                     bins=10,
                     color="blue",
                     alpha=0.5,
                        label="Blue"
                    
                    )
plt.legend()
plt.title("Weight Distribution",)
plt.xlabel("Weight")

plt.show()
```

```python
df=df[df["Win By"]!="Decision_Split"]
```

```python
plt.figure(figsize=(6,6),)
df["Win By"].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Win Method Distribution")
plt.ylabel("")
plt.show()
```

```python
top10=df["Fight Type"].value_counts().head(10)
sns.barplot(y=top10.index, x=top10.values,hue=top10.index,legend=False)
plt.title("Top 10 Fight Types")
plt.show()
```

```python


data = df.copy()

data["date"] = pd.to_datetime(data["date"])

years = data["date"].dt.year.value_counts().sort_index()

plt.figure(figsize=(10,5))

plt.plot(
                years.index,
                years.values,
                marker="o",
                color="blue",
            )

plt.title("UFC Fights Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Fights")

```

```python
wins=df["Winner"].value_counts().head(10)
sns.barplot(y=wins.index, x=wins.values,hue=wins.index,legend=False)
plt.title("Top 10 Winners")
plt.show()
```

4.DASHBOARD  COMPARISION

```python
plt.figure(figsize=(6,5))
sns.boxplot(data=df[["Red Age","Blue Age"]],palette=["red","blue"])
plt.title("Age Comparision by Corner")
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(data=df[["Red Reach","Blue Reach"]],palette=["red","blue"])
plt.title("Reach Comparision by Corner")
plt.show()
```

```python
weight_df=df[(df["Red Weight"]<=300) & (df["Blue Weight"]<=300)]

plt.figure(figsize=(6,5))
sns.boxplot(data=weight_df[["Red Weight","Blue Weight"]],palette=["red","blue"])
plt.title("Weight Comparision by Corner")
plt.show()
```

```python
plt.figure(figsize=(6,5))
plt.title("Height Comparision by Corner")
sns.boxplot(data=df[["Red Height","Blue Height"]],palette=["red","blue"])
plt.show()
```

```python
df["Age Diff"] = df["Red Age"] - df["Blue Age"]
df["Reach Diff"] = df["Red Reach"] - df["Blue Reach"]
df["Weight Diff"] = weight_df["Red Weight"] - weight_df["Blue Weight"]
df["Height Diff"] = df["Red Height"] - df["Blue Height"]
```

```python
compare=df[["Age Diff","Reach Diff","Weight Diff","Height Diff"]]
sns.heatmap(compare.corr(),annot=True,cmap="RdBu_r",center=0,fmt=".2f",)
plt.title("Correlation Heatmap")
plt.show()
```
